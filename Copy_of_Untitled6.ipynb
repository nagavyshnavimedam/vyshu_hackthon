```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
```

```python
np.random.seed(0)
hours_studied = np.random.randint(1,10,100)
pass_fail = np.where(hours_studied > 5,1,0)
```

```python
plt.scatter(hours_studied, pass_fail, colour='blue')
plt.xlabel('Hours Studied')
plt.ylabel('Pass(1)/Fail(0)')
plt.title('Hours Studied vs Pass/Fail')
plt.show()
```

```python
x_train, x_test,y_train,y_test = train_test_split(hours_studied.reshape(-1,1),pass_fail,test_size=0.3,random_state=42)
```

```python
model = LogisticRegression()
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
accuracy = accuracy_score(y_test,y_pred)
print(f"Accuracy:{accuracy}")
```

```python
plt.scatter(x_test,y_test,color='red')
plt.plot(np.sort(x_test,axis=0),model.predict_proba(np.sort(x_test,axis=0))[:,1],color='blue',linewidth=3)
plt.xlabel('Hours Studied')
plt.ylabel('Pass Probability')
plt.title('Hours Studied vs Pass Probability')
plt.show()
```
