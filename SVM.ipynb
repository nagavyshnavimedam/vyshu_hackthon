```python
selected_feature=['smoothness_mean,smoothness_worst']
```

```python
x=df['smoothness_mean']
y=df['smoothness_worst']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x.values.reshape(-1,1),y,test_size=0.2,random_state=38)
```

```python
model=SVR()
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
```

```python
mse=mean_squared_error(y_test,y_pred)
print("Mean Squared Error:",mse)
```
